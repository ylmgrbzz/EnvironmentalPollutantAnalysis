```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi içe aktarma
df = pd.read_csv('contaminants.csv')

# Gerekli olmayan sütunları kaldırma
df = df.drop(columns=['href', 'id'])

# NaN değerleri olan satırları düşürme
df = df.dropna()

# 'average_result' ve 'max_result' sütunlarını sayısal değerlere dönüştürme
df['average_result'] = pd.to_numeric(df['average_result'].str.extract(r'(\d+\.\d+)')[0], errors='coerce')
df['max_result'] = pd.to_numeric(df['max_result'].str.extract(r'(\d+\.\d+)')[0], errors='coerce')

# 'health_limit_exceeded' ve 'legal_limit_exceeded' sütunlarını ikili değerlere dönüştürme
df['health_limit_exceeded'] = df['health_limit_exceeded'].map({'Yes': 1, 'No': 0})
df['legal_limit_exceeded'] = df['legal_limit_exceeded'].map({'Yes': 1, 'No': 0})

# Veriyi özetleyen bazı temel analizler
summary_stats = df.describe(include='all')
print(summary_stats)

# Dağılım Grafiği (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='average_result', y='max_result', hue='health_limit_exceeded', data=df)
plt.title('Scatter Plot of Average Result vs Max Result')
plt.xlabel('Average Result')
plt.ylabel('Max Result')
plt.show()

# Çubuk Grafiği (Bar Plot) - Kontaminantlara göre ortalama sonuçlar
plt.figure(figsize=(15, 10))
avg_result_by_contaminant = df.groupby('contaminant')['average_result'].mean().sort_values(ascending=False)
sns.barplot(x=avg_result_by_contaminant.index, y=avg_result_by_contaminant.values)
plt.title('Average Result by Contaminant')
plt.xticks(rotation=90)
plt.xlabel('Contaminant')
plt.ylabel('Average Result')
plt.show()

# Histogram - Ortalama Sonuçların Dağılımı
plt.figure(figsize=(10, 6))
plt.hist(df['average_result'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Results')
plt.xlabel('Average Result')
plt.ylabel('Frequency')
plt.show()

# Korelasyon Matrisi ve Isı Haritası (Heatmap)
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Veriyi özetleyen bir tablo oluşturma
average_results = df['average_result'].dropna()
summary_table = average_results.describe()
print(summary_table)
```

```python
# Kategorik sütunlar arasında frekans sayımı (Count plot)
plt.figure(figsize=(15, 10))
sns.countplot(x='health_limit_exceeded', data=df)
plt.title('Count of Health Limit Exceeded')
plt.xlabel('Health Limit Exceeded')
plt.ylabel('Count')
plt.show()
```

```python
# Çubuk grafiği (Bar plot)
plt.figure(figsize=(15, 10))
df['contaminant'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Contaminants by Frequency')
plt.xlabel('Contaminant')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

Dağılım Grafiği (Scatter Plot)
average_result ve max_result arasındaki ilişkiyi inceleyerek hangi kirleticilerin daha tehlikeli olduğunu belirlemek

Çubuk Grafiği (Bar Plot)
En sık karşılaşılan kirletici maddeleri belirlemek ve hangi kirleticilerin daha yaygın olduğunu görmek.

Kutu Grafiği (Box Plot)
Her kirletici için average_result dağılımını görmek.


Isı Haritası (Heatmap)
Korelasyon matrisini görselleştirerek hangi değişkenlerin birbiriyle daha güçlü ilişkili olduğunu anlamak.


Karar:

Verilen veri setine dayanarak, "health_limit_exceeded" (sağlık limiti aşıldı) ve "legal_limit_exceeded" (yasal limit aşıldı) gibi göstergelerin, özellikle bazı kirleticiler için tehlikeli seviyelere ulaştığı gözlemlenmiştir. Özellikle contaminant sütununda en sık karşılaşılan kirleticilerin dikkatle izlenmesi ve azaltılması gerekmektedir. "average_result" ve "max_result" arasındaki ilişki, kirleticilerin yoğunluk seviyelerinin sağlık üzerindeki etkilerini daha iyi anlamamıza yardımcı olmaktadır.
Örneğin, "lead" (kurşun) gibi yaygın kirleticiler yüksek average_result değerlerine sahiptir ve bu da sağlık limitlerini aşmaktadır. Bu nedenle, bu kirleticilerin yönetimi ve kontrolü, halk sağlığını korumak için öncelikli olmalıdır.
Ayrıca, korelasyon matrisinde görüldüğü üzere, bazı değişkenler arasında güçlü ilişkiler bulunmaktadır. Özellikle "average_result" ve "max_result" arasında yüksek bir korelasyon gözlemlenmiştir, bu da kirleticilerin yoğunluk seviyelerinin dikkatle izlenmesi gerektiğini göstermektedir.
Bu karar ve gerekçeler, veriyi analiz ederek ve görselleştirerek elde edilen bulgulara dayanmaktadır. Bu nedenle, çevresel kirleticilerin izlenmesi ve yönetilmesi konusunda daha etkin önlemler alınması gerekmektedir.
